```python
import pandas as pd

df_list = pd.read_html('https://fbref.com/en/squads/53a2f082/2024-2025/all_comps/Real-Madrid-Stats-All-Competitions', attrs={"id":"stats_standard_combined"})
```

```python
player_stats_df = df_list[0]
print(player_stats_df.head())
```

```python
import os

# Get the current working directory (inside notebooks folder)
current_dir = os.getcwd()

# Navigate up to the main folder and then to the data folder
main_folder = os.path.dirname(current_dir)
data_folder = os.path.join(main_folder, "data")

# Ensure the data folder exists
os.makedirs(data_folder, exist_ok=True)

# Define the JSON file path
json_file_path = os.path.join(data_folder, "real_madrid_stats.json")

# Save the DataFrame as a JSON file
player_stats_df.to_json(json_file_path, orient="records", lines=True)

print(f"DataFrame has been saved to {json_file_path}")
```

```python
import json

# Define the path to the data folder from the notebooks folder
data_folder = os.path.join(os.pardir, "data")  # Navigate to parent folder and then into "data"
file_name = "Real_Madrid_players_market_value.json"
file_path = os.path.join(data_folder, file_name)

# Load the JSON file
with open(file_path, 'r') as file:
    market_value_data = json.load(file)

# Create a DataFrame from the JSON file
market_value_df = pd.json_normalize(market_value_data, record_path=['players'])

# Display the first few rows of the DataFrame to verify
print(market_value_df.head())
```

```python
# Display the columns of player_stats_df
player_stats_df_columns = player_stats_df.columns
print(player_stats_df_columns.tolist())
```

```python
# Rename and clean columns
player_stats_df.columns = [
    'Player', 'Nation', 'Position', 'Age', 'Matches_Played', 'Starts', 
    'Minutes_Played', '90s_Played', 'Goals', 'Assists', 'Goal_Contributions',
    'Non_Penalty_Goals', 'Penalty_Goals', 'Penalty_Attempts', 'Yellow_Cards', 
    'Red_Cards', 'Expected_Goals', 'Non_Penalty_xG', 'Expected_Assists', 
    'npxG_plus_xA', 'Progressive_Carries', 'Progressive_Passes', 
    'Progressive_Runs', 'Per90_Goals', 'Per90_Assists', 'Per90_Goal_Contributions',
    'Per90_Non_Penalty_Goals', 'Per90_Non_Penalty_Goal_Contributions', 
    'Per90_xG', 'Per90_xA', 'Per90_xG_plus_xA', 'Per90_Non_Penalty_xG', 
    'Per90_Non_Penalty_xG_plus_xA', 'Matches'
]

# Drop unnecessary columns (e.g., 'Matches' if redundant)
player_stats_df.drop(columns=['Matches', 'Per90_xG_plus_xA','Per90_Non_Penalty_xG', 'Per90_Non_Penalty_xG_plus_xA','Per90_Goal_Contributions', 'Per90_Non_Penalty_Goals', 'Per90_Non_Penalty_Goal_Contributions'], inplace=True)

# Display the cleaned DataFrame columns
print(player_stats_df.columns.tolist())
```

```python
print(player_stats_df.columns)
print(f"Number of columns: {len(player_stats_df.columns)}")
```

```python
# Drop columns not needed for analysis
columns_to_drop = ['npxG_plus_xA', 'Non_Penalty_xG']
player_stats_df.drop(columns=columns_to_drop, inplace=True)

# Display the updated DataFrame
print(player_stats_df.columns.tolist())
```

```python
print(player_stats_df.columns)
print(f"Number of columns: {len(player_stats_df.columns)}")
```

```python
print(market_value_df.columns.tolist())
```

```python
market_value_df['Player_Name'] = market_value_df['Player'].apply(
    lambda x: x[0] if isinstance(x, list) else x
).str.strip()
```

```python
print(market_value_df[['Player', 'Player_Name']].head())
```

```python
# Merging player_stats_df and market_value_df dataframes
merged_df = pd.merge(player_stats_df, market_value_df, left_on='Player', right_on='Player_Name', how='inner')

# Display the first few rows of the merged DataFrame
print(merged_df.head())
```

```python
# Drop unnecessary columns
columns_to_drop = ['Player_y', 'Age', 'Signed from', 'Joined', 'Nation', 'Player_Name']
merged_df_cleaned = merged_df.drop(columns=columns_to_drop)

# Rename Player_x to Player for simplicity
merged_df_cleaned.rename(columns={'Player_x': 'Player'}, inplace=True)
```

```python
merged_df_cleaned
```

```python
# Check if there are any NaN values in each column
nan_columns = merged_df_cleaned.isnull().any()
print(nan_columns)
```

```python
# Replace NaN values with 0 in the entire DataFrame
merged_df_cleaned.fillna(0, inplace=True)

# Confirm there are no NaN values left
print("NaN values handled successfully. Any NaN remaining?")
print(merged_df_cleaned.isnull().any().any())  # This should return False
```

```python
# Define the path to the data folder
data_folder_path = os.path.join(os.pardir, "data")
cleaned_file_name = "merged_player_stats_market_value.csv"
cleaned_file_path = os.path.join(data_folder_path, cleaned_file_name)

# Save the cleaned DataFrame as a CSV file in the data folder
merged_df_cleaned.to_csv(cleaned_file_path, index=False)

print(f"The cleaned CSV file has been successfully saved at: {cleaned_file_path}")
```
